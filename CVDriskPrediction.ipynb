# Data Pre-Processing

## Cleaning the Data

```python
# Calculate the mode of the column
mode_value = cvd['education'].mode().iloc[0]  

# Fill null values with the mode
cvd['education'] = cvd['education'].fillna(mode_value)  
```

```python
# Calculate the mode of the column
mode_value_BPMeds = cvd['BPMeds'].mode().iloc[0]  

# Fill null values with the mode (common value)
cvd['BPMeds'] = cvd['BPMeds'].fillna(mode_value)  
cvd['BPMeds'] = cvd['BPMeds'].astype(int)
```

```python
# Fill missing values in cigsPerDay, glucose, heartRate and totChol column with mean value (average value)
cvd['cigsPerDay'].fillna(cvd['cigsPerDay'].mean(), inplace=True)
cvd['totChol'].fillna(cvd['totChol'].mean(), inplace=True)
cvd['glucose'].fillna(cvd['glucose'].mean(), inplace=True)
cvd['heartRate'].fillna(cvd['heartRate'].mean(), inplace=True)

# Fill missing values in BMI (midpoint value)
cvd['BMI'].fillna(cvd['BMI'].median(), inplace=True)
```

```python
# making sure the missing values are replaced and are no longer present in the dataset
print("Checking for any NaN values after replacing them in the dataset :",'\n', cvd.isna().sum())
```

```python
# dropping the 'non-existent' duplicate value
cvd.drop_duplicates(inplace=True)

# displaying the dataset shape after deleting duplicated values to avoid data leakage
print("Dataset shape after deleting all duplicate values: ", cvd.shape, '\n')
```

```python
# remove columns of id, considered irrelevant
cvd.drop(['id'], axis=1, inplace=True)
print("Dataset after deleting id values: ")
cvd.head
```

```python
# check for categorical variables
categorical_columns = cvd.select_dtypes(include=['object']).columns
print("\nCategorical Columns:")
print(categorical_columns)
```

```python
# import necessary libraries
from sklearn.preprocessing import LabelEncoder

# encoding using label encoder 
le = LabelEncoder()

# categorical features
categorical_columns = ['sex', 'is_smoking']

for column in categorical_columns:
    cvd[column] = le.fit_transform(cvd[column])
```

## Handle Outliers

```python
# Function to detect and print outliers using IQR method
def detect_outliers(cvd, column):
    Q1 = cvd[column].quantile(0.25)
    Q3 = cvd[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = cvd[(cvd[column] < lower_bound) | (cvd[column] > upper_bound)]
    
    return outliers.shape[0]

# Sum up the outliers for selected columns
total_outliers = 0

# Define columns with outliers
columns_with_outliers = ['education', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']

for column in columns_with_outliers:
    outliers_count = detect_outliers(cvd, column)
    total_outliers += outliers_count
    print(f"Number of outliers in {column}: {outliers_count}")

print("\nTotal number of outliers in all columns:", total_outliers)
```

```python
# Function to handle outliers by replacing them with the mean
def handle_outliers_with_median(cvd, column):
    Q1 = cvd[column].quantile(0.25)
    Q3 = cvd[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with the median
    cvd[column] = np.where((cvd[column] < lower_bound) | (cvd[column] > upper_bound), 
                           cvd[column].median(), cvd[column])
    
     # Return the count of outliers
    outliers_count = cvd[(cvd[column] < lower_bound) | (cvd[column] > upper_bound)].shape[0]
    return outliers_count if outliers_count > 0 else 0
```

```python
# Handle outliers for selected columns
for column in columns_with_outliers:
    handle_outliers_with_median(cvd, column)

# Visualize the changes using boxplots
plt.figure(figsize=(12, 8))

for i, column in enumerate(columns_with_outliers, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(x='TenYearCHD', y=column, data=cvd)

plt.suptitle("Boxplots after Handling Outliers with Median", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Handle outliers for selected columns and print the count
total_outliers_after_median = 0

for column in columns_with_outliers:
    outliers_count = handle_outliers_with_median(cvd, column)
    total_outliers_after_median += outliers_count
    print(f"Number of outliers in {column} after replacing with median: {outliers_count}")

print("\nTotal number of outliers in all columns after replacing with median:", total_outliers_after_median)
```

### Scaling the Data 

```python
cvd_scaled = cvd_unscaled
X_scaled = cvd_scaled.iloc[:,:-1]
y_scaled = cvd_scaled.iloc[:,-1]
print(X_scaled)
print(X_scaled.shape)

# scaling the data
def scaled_data(cvd_x):
    scale.fit(cvd_x)
    X_scaled = scale.transform(cvd_x)
    return X_scaled

X_scaled = scaled_data(X_scaled)
print(X_scaled)
print(X_scaled.shape)

X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_scaled,y_scaled, test_size=0.25, random_state= 42)
```

## Model Sampling 

```python
# importing necessary libraries for balancing of dataset
from imblearn.over_sampling import SMOTE

# applying SMOTE to balance the dataset
smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X_scaled, y_scaled)

# Splitting the resampled data into training and testing sets
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42)
```

## Applying ML Algorithms 

```python
from sklearn import metrics
from sklearn.metrics import classification_report
import time

def fitTest(model, show_score):
    start = time.time()
    
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    end = time.time()
    time_taken = end-start
    if show_score:
        print("Using: ", str(model))
        print("Training score: {:.3f}".format(model.score(X_train, y_train)))
        print("Accuracy score: {:.3f}".format(metrics.accuracy_score(y_test, y_pred)))
        print(classification_report(y_test, y_pred))
        print("Time taken to complete: {:.3f} seconds".format(time_taken))

def scaledTest(model, show_score):
    start = time.time()
    
    model.fit(X_train_scaled,y_train_scaled)
    y_pred_scaled = model.predict(X_test_scaled)
    
    end = time.time()
    time_taken = end-start
    if show_score:
        print("Using: ", str(model))
        print("Training score for scaled and balanced data: {:.3f}".format(model.score(X_train_scaled, y_train_scaled)))
        print("Accuracy score for scaled and balanced data: {:.3f}".format(metrics.accuracy_score(y_test_scaled, y_pred_scaled)))
        print(classification_report(y_test_scaled, y_pred_scaled))
        print("Time taken to complete: {:.3f} seconds".format(time_taken))
        
```

# Model Tuning - Hyperparameters

```python
# This technique randomly samples from a grid of hyperparameters instead
# of conducting an exhaustive search (Selvaraj, 2022)
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
```

## Logistic Regression

```python
# Start measuring time
start_time = time.time()

# Define hyperparameters for the model
param_dist_logreg = {
    'penalty': ['l1', 'l2'],
    'C': uniform(0.1, 10),  # Use a continuous distribution for the regularization parameter C
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 300, 400, 500],
}

# Apply random search for the model
random_search_logreg = RandomizedSearchCV(LogisticRegression(), param_distributions=param_dist_logreg, n_iter=10, random_state=42)

# Fit the model with the best hyperparameters
random_search_logreg.fit(X_train_scaled, y_train_scaled)

# Stop measuring time
end_time = time.time()

# Calculate the time taken
time_taken = end_time - start_time

# Get the best hyperparameters for the model
best_params_logreg = random_search_logreg.best_params_

# Display the best hyperparameters and time taken
print("Best Hyperparameters - Logistic Regression:", best_params_logreg)
print("Time taken to complete: {:.3f} seconds".format(time_taken))
```

##  MLP Classifier

```python
# Start measuring time
start_time = time.time()

# Define hyperparameters for the model
param_dist_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50, 25)],
    'activation': ['relu', 'tanh', 'logistic'],
    'solver': ['sgd', 'adam'],
    'alpha': uniform(0.0001, 1),  # Use a continuous distribution for alpha
}

# Apply random search for the model
random_search_mlp = RandomizedSearchCV(MLPClassifier(), param_distributions=param_dist_mlp, n_iter=10, random_state=42)

# Fit the model with the best hyperparameters
random_search_mlp.fit(X_train_scaled, y_train_scaled)

# Stop measuring time
end_time = time.time()

# Calculate the time taken
time_taken = end_time - start_time

# Get the best hyperparameters for the model
best_params_mlp = random_search_mlp.best_params_

# Display the best hyperparameters and time taken
print("Best Hyperparameters - MLPClassifier:", best_params_mlp)
print("Time taken to complete: {:.3f} seconds".format(time_taken))
```
